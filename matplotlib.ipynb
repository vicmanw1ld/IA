REPRESENTACIÓN GRÁFICA DE DATOS

Si a la función de trazado se le da una matriz de datos, la usará como coordenadas en el eje vertical, y utilizará el índice de cada punto de datos en el array como la coordenada horizontal.

```python
plt.plot([1, 2, 5, 7, 8, 3, 1]) #FUNCIÓN plot
plt.show()
```

También se puede proporcionar dos matrices: una para el eje horizontal, y otra para el eje vertical.

```python
#para ello le damos dos listas que una será el eje x y la otra el eje y
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.show()
```

Pueden modificarse las logitudes de los ejes para que la figura no se vea tan ajustada

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.axis([-4, 8, 0, 10]) # [xmin, xmax, ymin, ymax] esto son los ejes, para darlo más margen a los bordes
plt.show()
```

Se sigue el mismo procedimiento para pintar una función matemática

```python
import numpy as np
x = np.linspace(-2, 2, 500) # esto genera un array con linspace y va del -2 al 2 y tiene 500 valores
y = x**2 # y esta es la función, de X al cuadrado **

plt.plot(x, y)
plt.show()
x # este es el array de 500 valores de -2 a 2
```

También pude modificarse el estilo de la gráfica para que contenga más información. Como ponerles nombres a los vertices, a la tabla o poner cuadrícula por dentro

```python
plt.plot(x, y)
plt.title("TITULO: Square function")
plt.xlabel("x - EJE HORIZONTAL") # ETIQUETA EN EL EJE HORIZONTAL
plt.ylabel("y = x**2 EJE VERTICAL") #Etiqueta en el eje vertical
plt.grid(True) # El grid a True es la cuadricula
plt.show()
```

Pueden superponerse gráficas y cambiar el estilo de las líneas

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1 # OTRA FUNCIÓN

plt.plot(x, y, 'b--', x, y2, 'g') #Puedo superponer y ademas 'b - - ' es blue discontinua y g green continua
plt.show()
```

Es mejor colocarlo así

```python
# Separando en diferentes lineas las funciones
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--')
plt.plot(x, y2, 'g')
plt.show()
```

Para poder diferenciar entre ambas funciones siempre es recomendable añadir una leyenda

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--', label="x**2") # le ponemos la linea y la etiqueta de la 1 función
plt.plot(x, y2, 'g', label="x+1") # le ponemos la linea y la etiqueta de la 2 función
plt.legend(loc="best") # La leyenda la situa en la mejor localización donde el considere que queda mejor
plt.show()
```

Tambien puede crearse dos graficas que no se superpongan. Estas graficas se organizan en un grid y se denominan subplots.

```python
# lo podemos añadir en dos cuadros diferentes
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.subplot(1, 2, 1) # 1 rows, 1 columns, 1st subplot, numero de celdas, de columnas y te lo pone en el 1er cuadro
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) # 1 rows, 2 columns, 2nd subplot, aqui te hace 1 celda, 2 columnas y te lo pone en el 2º cuadro
plt.plot(x, y2, 'g')

plt.show()
```

```python
#Para que las gráficas no queden tan ajustadas, podemos hacer la figura más grande.
plt.figure(figsize=(14,6)) # funcion para aumentar el tamaño figsize

plt.subplot(1, 2, 1) # 1 rows, 1 columns, 1st subplot
plt.plot(x, y, 'b--')
plt.xlabel("x1", fontsize=14) # Para aumentar la letra
plt.ylabel("y1", fontsize=14)
plt.title("TITULO")
plt.xlabel("x - EJE HORIZONTAL") # ETIQUETA EN EL EJE HORIZONTAL
plt.ylabel("y - EJE VERTICAL") #Etiqueta en el eje vertical
plt.grid(True) # El grid a True es la cuadricula

plt.subplot(1, 2, 2) # 1 rows, 2 columns, 2nd subplot
plt.plot(x, y2, 'g')
plt.xlabel("x2", fontsize=14)
plt.ylabel("y2", fontsize=14)
plt.title("TITULO")
plt.xlabel("x - EJE HORIZONTAL") # ETIQUETA EN EL EJE HORIZONTAL
plt.ylabel("y - EJE VERTICAL") #Etiqueta en el eje vertical
plt.grid(True) # El grid a True es la cuadricula

plt.show()
#podemos ponerle todos los nombres y demás pero a cada una de ellas
```

SCATTER PLOTS (GRAFICAS CON PUNTOS)

```python
from numpy.random import rand

x, y = rand(2, 100) #VALORES ALEATORIOS ENTRE 2 Y 100
plt.scatter(x, y)
plt.show()

```

```python
from numpy.random import rand

x, y = rand(2, 100)
x2, y2 = rand(2, 100)

plt.scatter(x, y, c='red') #le indico el color también

plt.scatter(x2, y2, c='blue')

plt.show()
```

HISTOGRAMA Son diferentes barritas que suben a un valor determinado

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]

plt.subplot(211) # 2 rows, 1 columns, 1st subplot, hace dos gráficos
plt.hist(data, bins = 15, rwidth=0.8) #funcion hist, le pasa bins (son los intervalos) y el ancho de las barras 0.8
plt.xlabel("Value") #etiquetas
plt.ylabel("Frequency") # etiqueta

plt.subplot(212) # 2 rows, 1 columns, 2nd subplot
plt.hist(data, bins = [1, 1.5, 2, 2.5, 3], rwidth=0.95) # aqui le cambia los bins
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
# el 3 que se repite 4 veces vemos en la 1ª grafica como sube, se incrementa
```
