```python
import pandas as pd
```

CREACIÓN DEL OBJETO SERIES

```python
#creación del objeto series
s = pd.Series([2,4,6,8,10])
s #lo de la izda es un índice y a la derecha lo que hemos puesto
```

```python
# Creación de un objeto Series inicializándolo con un diccionario de Python
altura = {"Santiago": 187, "Pedro": 178, "Julia": 170, "Ana": 165}
s = pd.Series(altura)
print(s) #me crea el indice con los nombres
```

```python
# Creación de un objeto Series inicializándolo con algunos
# de los elementos de un diccionario de Python
altura = {"Santiago": 187, "Pedro": 178, "Julia": 170, "Ana": 165}
s = pd.Series(altura, index = ["Pedro", "Julia"])
print(s)
```

```python
# Creación de un objeto Series inicializandolo con un escalar
s = pd.Series(34, ["test1", "test2", "test3"])
print(s)
```

ACCESO AL OBJETO SERIES

```python
#Cada elemento en un objeto Series tiene un identificador único que se denomina **_index label_**.
# Creación de un objeto Series
s = pd.Series([2, 4, 6, 8], index=["num1", "num2", "num3", "num4"])
print(s)
```

```python
# Accediendo al tercer elemento del objeto
s["num3"]
# Nos da el elemento 6, tiene mejores características pq puedes acceder a el por el numero de posición
s[2]
```

```python
# loc es la forma estándar de acceder a un elemento de un objeto Series por atributo
s.loc["num3"]
#lo hacemos cuando queremos acceder a un elemento por el indice con loc
```

```python
# iloc es la forma estándar de acceder a un elemento de un objeto Series por posición
s.iloc[2]
```

```python
# Accediendo al segundo y tercer elemento por posición
s.iloc[2:4]
```

Operaciones aritméticas con Series

```python
# Creacion de un objeto Series
s = pd.Series([2, 4, 6, 8, 10])
print(s)
```

```python
# Los objeto Series son similares y compatibles con los Arrays de Numpy
import numpy as np
# Ufunc de Numpy para sumar los elementos de un Array
np.sum(s) #me suma todos los números del objeto series
# El resto de operaciones aritméticas de Numpy sobre Arrays también son posibles
# Más información al respecto en la Introducción a Numpy
s * 2
```

CREACIÓN DE UN OBJETO DATAFRAME

```python
# Creación de un DataFrame inicializándolo con un diccionario de objetios Series
personas = {
    "peso": pd.Series([84, 90, 56, 64], ["Santiago","Pedro", "Ana", "Julia"]),
    "altura": pd.Series({"Santiago": 187, "Pedro": 178, "Julia": 170, "Ana": 165}),
    "hijos": pd.Series([2, 3], ["Pedro", "Julia"])
}

df = pd.DataFrame(personas)
df
#NaN es un valor nulo o que no existe
```

Puede forzarse al DataFrame a que presente unas columnas determinadas y en un orden determinado

```python
# Creación de un DataFrame inicializándolo con algunos elementos de un diccionario
# de objetos Series
personas = {
    "peso": pd.Series([84, 90, 56, 64], ["Santiago","Pedro", "Ana", "Julia"]),
    "altura": pd.Series({"Santiago": 187, "Pedro": 178, "Julia": 170, "Ana": 165}),
    "hijos": pd.Series([2, 3], ["Pedro", "Julia"])
}

df = pd.DataFrame(personas, columns = ["altura", "peso"], index = ["Ana", "Julia", "Santiago"])
df
# VER LOS MÉTODOS QUE TIENE Y JUGAR UN POCO
```

| Método          | Descripción                                  |
| --------------- | -------------------------------------------- |
| `df.head()`     | Muestra las primeras 5 filas                 |
| `df.tail()`     | Muestra las últimas 5 filas                  |
| `df.info()`     | Información general (tipos, nulos, memoria)  |
| `df.shape`      | Muestra el número de filas y columnas        |
| `df.columns`    | Nombres de las columnas                      |
| `df.index`      | Índices del DataFrame                        |
| `df.describe()` | Estadísticas resumen para columnas numéricas |
| `df.dtypes`     | Tipos de datos por columna                   |

| Método                                   | Descripción                    |
| ---------------------------------------- | ------------------------------ |
| `df["col"]`                              | Accede a una columna           |
| `df.loc["Ana"]`                          | Accede a una fila por etiqueta |
| `df.iloc[0]`                             | Accede a una fila por posición |
| `df["nueva_col"] = ...`                  | Agrega una columna             |
| `df.drop(columns=["peso"])`              | Elimina columnas               |
| `df.drop(index=["Pedro"])`               | Elimina filas                  |
| `df.rename(columns={"altura": "talla"})` | Renombra columnas              |

| Método                         | Descripción                      |
| ------------------------------ | -------------------------------- |
| `df.mean()`                    | Media de cada columna numérica   |
| `df.sum()`                     | Suma de cada columna             |
| `df.max()` / `df.min()`        | Máximo y mínimo                  |
| `df.apply()`                   | Aplica una función personalizada |
| `df.sort_values("peso")`       | Ordena por columna               |
| `df.isnull()` / `df.notnull()` | Localiza valores nulos           |

| Método         | Descripción                     |
| -------------- | ------------------------------- |
| `df.fillna(0)` | Rellena nulos con 0             |
| `df.dropna()`  | Elimina filas con valores nulos |



```python
# vamos a hacer el IMC añadiendo una columna IMC = peso/altura en m ^2
df["IMC"] = df["peso"] / ((df["altura"] / 100) ** 2)
df
#Ahora lo vamos a clasificar por ESTADO incluyendo el IMC
def clasificar_imc(imc):
    if imc < 18.5:
        return "delgado"
    elif imc < 25:
        return "medio"
    else:
        return "gordo"

df["estado"] = df["IMC"].apply(clasificar_imc)
df
```

```python
#filtrar por altura
altos = df[df["altura"] > 170]
altos
```

```python
# El que más pesa, con la función idxmax
persona_mas_peso = df["peso"].idxmax()
peso_mas_alto = df["peso"].max()

print(f"La persona con más peso es {persona_mas_peso} con {peso_mas_alto} kg.")
```

```python
#Se pueden usar funciones de NUMPY con el df
peso_max = np.max(df["peso"])
peso_min = np.min(df["peso"])
peso_medio = np.mean(df["peso"])
peso_std = np.std(df["peso"])

print("Peso máximo:", peso_max)
print("Peso mínimo:", peso_min)
print("Peso medio:", peso_medio)
print("Desviación estándar:", peso_std)
```
